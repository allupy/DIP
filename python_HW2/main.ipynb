```python
# 直方图均衡
import cv2
import os
import matplotlib.pyplot as plt
from myfuntion import(
    myHisteq
)


filenames = ['bridge','circuit','moon']
for filename in filenames:
    print(filename)

    # 读取图像
    im = cv2.imread(f'{filename}.jpg')    
    
    # if len(im.shape) == 2:  # 灰度图  
    #     print("0")  
    # elif len(im.shape) == 3 and im.shape[2] == 1:  # 在某些情况下，OpenCV可能以(高度, 宽度, 1)的形式返回灰度图  
    #     print("2")  
    # elif len(im.shape) == 3 and im.shape[2] == 3:  # RGB图  
    #     print("3")  
    # else:  
    #     print("这是其他类型的图像")

    n_1 = 2
    n_2 = 64
    n_3 = 256

    Histeq_1 = myHisteq(im,n_1)
    Histeq_2 = myHisteq(im,n_2)
    Histeq_3 = myHisteq(im,n_3)

    # print(f'n_1:{n_1},lut_1:{Histeq_1[1]}')
    # print(f'n_2:{n_2},lut_1:{Histeq_2[1]}')

    # 创建 result 文件夹
    if not os.path.exists('result'):
        os.makedirs('result')

    # 将结果保存到当前目录下的result文件夹下
    cv2.imwrite(f'result/_{filename}_{n_1:.1f}_Histogram Equalization.jpg', Histeq_1)  
    cv2.imwrite(f'result/_{filename}_{n_2:.1f}_Histogram Equalization.jpg', Histeq_2)
    cv2.imwrite(f'result/_{filename}_{n_3:.1f}_Histogram Equalization.jpg', Histeq_3)

    # 注意：matplotlib 默认使用 RGB 色彩空间，而 OpenCV 使用的是 BGR。  
    # 因此，在显示之前，我们需要将图片从 BGR 转换为 RGB。  
    im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)  
    Histeq_1_rgb = cv2.cvtColor(Histeq_1, cv2.COLOR_BGR2RGB)  
    Histeq_2_rgb = cv2.cvtColor(Histeq_2, cv2.COLOR_BGR2RGB)  
    Histeq_3_rgb = cv2.cvtColor(Histeq_3, cv2.COLOR_BGR2RGB) 
    # # 将结果保存到当前目录下的result文件夹下
    # cv2.imwrite(f'result/_{filename}_{n_1:.1f}_Histogram Equalization.jpg', Histeq_1[1])  
    # cv2.imwrite(f'result/_{filename}_{n_2:.1f}_Histogram Equalization.jpg', Histeq_2[1])
    # cv2.imwrite(f'result/_{filename}_{n_3:.1f}_Histogram Equalization.jpg', Histeq_3[1])

    # # 注意：matplotlib 默认使用 RGB 色彩空间，而 OpenCV 使用的是 BGR。  
    # # 因此，在显示之前，我们需要将图片从 BGR 转换为 RGB。  
    # im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)  
    # Histeq_1_rgb = cv2.cvtColor(Histeq_1[1], cv2.COLOR_BGR2RGB)  
    # Histeq_2_rgb = cv2.cvtColor(Histeq_2[1], cv2.COLOR_BGR2RGB)  
    # Histeq_3_rgb = cv2.cvtColor(Histeq_3[1], cv2.COLOR_BGR2RGB)  
    
    # 创建一个2x2的子图网格  
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))  
    
    # 将图片添加到子图中  
    axes[0, 0].imshow(im_rgb)  
    axes[0, 0].set_title('Original Image')  
    axes[0, 0].axis('off')  # 关闭坐标轴  
    
    axes[0, 1].imshow(Histeq_1_rgb)  
    axes[0, 1].set_title(f'Histogram Equalization\nn={n_1:.1f}')  
    axes[0, 1].axis('off')  
    
    axes[1, 0].imshow(Histeq_2_rgb)  
    axes[1, 0].set_title(f'Histogram Equalization\nn={n_2:.1f}')  
    axes[1, 0].axis('off')  
    
    axes[1, 1].imshow(Histeq_3_rgb)  
    axes[1, 1].set_title(f'Histogram Equalization\nn={n_3:.1f}')  
    axes[1, 1].axis('off')  
    
    # 调整子图之间的间距并显示图表  
    plt.tight_layout()  
    plt.savefig(f'result/{filename}_combined_histogram_equalization.jpg')
    plt.show()


    #plt.savefig(f'result/{filename}_combined_histogram_equalization.jpg')

    # plt.show()
```

```python
# 均值滤波
import cv2
import os
import matplotlib.pyplot as plt
from myfuntion import(
    myAverage
)


filenames = ['circuit']
for filename in filenames:
    print(filename)

    # 读取图像
    im = cv2.imread(f'{filename}.jpg')    
    n_1 = 1
    n_10 = 10
    n_100 = 100
    im_1 = myAverage(im)
    im_10 = myAverage(im)
    im_100 = myAverage(im)
    for i in range(n_1-1):
        im_10 = myAverage(im_10)

    for i in range(n_100-1):
        im_100 = myAverage(im_100)

    # 创建 result 文件夹
    if not os.path.exists('result'):
        os.makedirs('result')

    # 将结果保存到当前目录下的result文件夹下
    #cv2.imwrite(f'result/_{filename}_{n_1:.1f}_Histogram Equalization.jpg', im_a)  


    # 注意：matplotlib 默认使用 RGB 色彩空间，而 OpenCV 使用的是 BGR。  
    # 因此，在显示之前，我们需要将图片从 BGR 转换为 RGB。  
    im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)  
    Histeq_1_rgb = cv2.cvtColor(im_1, cv2.COLOR_BGR2RGB)  
    Histeq_2_rgb = cv2.cvtColor(im_10, cv2.COLOR_BGR2RGB)
    Histeq_3_rgb = cv2.cvtColor(im_100, cv2.COLOR_BGR2RGB)

    
    # 创建一个2x2的子图网格  
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))  
    
    # 将图片添加到子图中  
    axes[0, 0].imshow(im_rgb)  
    axes[0, 0].set_title('Original Image')  
    axes[0, 0].axis('off')  # 关闭坐标轴  
    
    axes[0, 1].imshow(Histeq_1_rgb)  
    axes[0, 1].set_title(f'Average Filter image\nn= {n_1}')  
    axes[0, 1].axis('off')  
    
    axes[1, 0].imshow(Histeq_2_rgb)  
    axes[1, 0].set_title(f'Average Filter image\nn={n_10}')  
    axes[1, 0].axis('off')  
    
    axes[1, 1].imshow(Histeq_3_rgb)  
    axes[1, 1].set_title(f'Average Filter image\nn={n_100}')  
    axes[1, 1].axis('off')  
    
    # 调整子图之间的间距并显示图表  
    plt.tight_layout()  
    plt.savefig(f'result/{filename}_combined_filter.jpg')
    plt.show()


    #plt.savefig(f'result/{filename}_combined_histogram_equalization.jpg')

    # plt.show()
```

```python
#图像锐化
import cv2
import os
import matplotlib.pyplot as plt
from myfuntion import(
    mySharpen
)


filenames = ['moon']
for filename in filenames:
    print(filename)

    # 读取图像
    im = cv2.imread(f'{filename}.jpg') 
    n_1 = 1
    n_2 = 0.5
    n_3 = 1.5   
    
    sharpen_1 = mySharpen(im,n_1)
    sharpen_2 = mySharpen(im,n_2)
    sharpen_3 = mySharpen(im,n_3)



    # 创建 result 文件夹
    if not os.path.exists('result'):
        os.makedirs('result')


    # 注意：matplotlib 默认使用 RGB 色彩空间，而 OpenCV 使用的是 BGR。  
    # 因此，在显示之前，我们需要将图片从 BGR 转换为 RGB。  
    im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)  
    sharpen_1 = cv2.cvtColor(sharpen_1, cv2.COLOR_BGR2RGB)  
    sharpen_2 = cv2.cvtColor(sharpen_2, cv2.COLOR_BGR2RGB)
    sharpen_3 = cv2.cvtColor(sharpen_3, cv2.COLOR_BGR2RGB)
    # 创建一个1x2的子图网格  
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))  
    
    # 将图片添加到子图中  
    axes[0, 0].imshow(im_rgb)  
    axes[0, 0].set_title('Original Image')  
    axes[0, 0].axis('off')  # 关闭坐标轴  
    
    axes[0, 1].imshow(sharpen_1)  
    axes[0, 1].set_title(f'sharpen\nalpah={n_1:.1f}')  
    axes[0, 1].axis('off')  
    
    axes[1, 0].imshow(sharpen_2)  
    axes[1, 0].set_title(f'sharpen\nalpah={n_2:.1f}')  
    axes[1, 0].axis('off')  
    
    axes[1, 1].imshow(sharpen_3)  
    axes[1, 1].set_title(f'sharpen\nalpah={n_3:.1f}')  
    axes[1, 1].axis('off')   
    
    
    # 调整子图之间的间距并显示图表  
    plt.tight_layout()  
    plt.savefig(f'result/{filename}_combined_sharpen.jpg')
    plt.show()

```
